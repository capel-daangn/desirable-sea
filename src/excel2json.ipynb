```python
import pandas as pd

df_origin = pd.read_csv("제주관광공사_마을편의시설_20220930.csv")
```

```python
display(df_origin.info())
display(df_origin.head())
```

```python
df_extracted = df_origin[["편의시설명","편의시설유형","소재지도로명주소","위도", "경도"]]
df_extracted.head()
```

```python
df_parking_lot = df_extracted[df_extracted["편의시설유형"] == "주차장"]
df_parking_lot.head()
```

```python
df_parking_lot.info()
```

```python
df_parking_lot.columns = ['name', 'type', 'address', 'lat', 'lng']
```

```python
df_parking_lot.to_json("data.json", orient = 'records', force_ascii=False)
```
